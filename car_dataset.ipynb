1.Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

2.Load the dataset using Pandas read_csv() function.

```python
car=pd.read_csv('/content/drive/MyDrive/cars.csv')
car
```

3.Display the head of the dataset using the head() function.

```python
car.head()
```

4.Display the bottom 5 rows from the dataset using the tail() function.

```python
car.tail(5)
```

5.Print summary statistics of the dataset using the describe() function.

```python
car.describe(include='all')
```

6.Plot a histogram for all the variables.

```python
car.hist(figsize=(20,30))
```

7.Box plot to visualize the relationship between vehicle size and engine hp.

```python
sns.boxplot(x='mpg',y='disp',data=car)
```

8.Build a pair plot using the seaborn library.

```python
sns.pairplot(car)
```

9.Drop irrelevant columns from the dataset using drop() function.

```python
car=car.drop(['vs','am'],axis=1)
car.head(5)
```

10.Use rename() function to rename the columns.

```python
car=car.rename(columns={"mpg":"MPG","hp":"HP"})
car.head(5)
```

11.Print the total number of duplicate rows.

```python
duplicate_rows_car=car[car.duplicated()]
print("Number of dulicate rows:",duplicate_rows_car.shape)
```

```python
car.count()
```

12.Remove the duplicate rows using the drop_duplicate() function.

```python
car=car.drop_duplicates()
car.head()
```

13.Drop the missing values from the dataset.

```python
print(car.isnull().sum())
```

```python
car=car.dropna()
car.count()
```

14.Plot a histogram to find the number of cross per brand.

```python
car.drat.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by drat")
plt.ylabel("Number of cars")
plt.xlabel("drat");
```

```python
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
